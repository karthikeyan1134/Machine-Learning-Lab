```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Load the dataset
tennis_data = pd.read_csv('/content/sample_data/play_tennis.csv')

# Separate features (X) and target (y)
features = tennis_data.iloc[:, :-1]
target = tennis_data.iloc[:, -1]

# Label encode the features and target
label_encoder = LabelEncoder()
encoded_features = features.apply(label_encoder.fit_transform)
encoded_target = label_encoder.fit_transform(target)

# Initialize and train the decision tree classifier
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(encoded_features, encoded_target)

# Plot the decision tree
plt.figure(figsize=(12, 8))  # Adjust figure size for better readability
tree.plot_tree(decision_tree_model, filled=True, feature_names=features.columns, class_names=label_encoder.classes_, rounded=True)
plt.title('Decision Tree Representation')
plt.show()
```
